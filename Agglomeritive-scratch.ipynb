```python
def distance(x,y):
    return abs((x[0]-y[0])+(x[1]-y[1]))
```

```python
def single(l,m):
    return min([distance(i,j) for i in l for j in m])
def complete(l,m):
    return max([distance(i,j) for i in l for j in m])
def average(l,m):
    return np.mean([distance(i,j) for i in l for j in m])
def centroid(l,m):
    return np.median([distance(i,j) for i in l for j in m])
```

```python
def agglo(data, method):
    clusters = [[point] for point in data]

    while len(clusters) > 1:
        dic = {}
        for i, cluster1 in enumerate(clusters):
            for j, cluster2 in enumerate(clusters):
                if i != j: 
                    if method == "single":
                        dic[(i, j)] = single(cluster1, cluster2)
                    elif method == "complete":
                        dic[(i, j)] = complete(cluster1, cluster2)
                    elif method == "average":
                        dic[(i, j)] = average(cluster1, cluster2)
                    elif method == "centroid":
                        dic[(i, j)] = centroid(cluster1, cluster2)

        mini=100
        for i in dic.items():
          if dic[i[0]]<mini and dic[i[0]]!=0:
            mini=dic[i[0]]
            name=i[0]
        closest_pair = name
       # print(closest_pair)
        mini = dic[closest_pair]
        print(f"Minimum distance: {mini}")
        print(f"Closest pair: {closest_pair}")

        new_cluster = clusters[closest_pair[0]] + clusters[closest_pair[1]]
        clusters = [cluster for i, cluster in enumerate(clusters) if i not in closest_pair]
        clusters.append(new_cluster)

        print(f"Updated clusters: {clusters}")
        print(f"Number of clusters: {len(clusters)}\n")
```

```python
points = [
    (1, 1), (2, 2), (3, 1), (12, 9), (6, 1), (9, 1),
    (11, 1), (13, 9), (1, 9), (1, 10), (1, 11), (12, 10),
    (3, 9), (3, 10), (3, 11), (13, 10), (4, 11), (6, 6),
    (6, 7), (14, 11), (6, 8), (7, 6), (7, 7), (8, 7),
    (13, 5), (14, 5), (13, 6), (10, 8), (11, 8)
]

methods = ['complete', 'average', 'centroid','single']
for method in methods:
    print(f"\nAgglomerative Clustering with {method.capitalize()} Linkage:\n")
    agglo(points, method)
```
