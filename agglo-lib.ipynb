```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

data = np.array([ [1, 1], [2, 2], [3, 1],[12, 9],[6, 1],[9, 1],[11, 1],[13, 9],
                     [1, 9],[1, 10],[1, 11],[12, 10],[3, 9],[3, 10],[3, 11],[13, 10],
                     [4, 11],[6, 6],[6, 7],[14, 11],[6, 8],[7, 6],[7, 7],
                     [8, 7],[13, 5],[14, 5],[13, 6],[10, 8],[11, 8] ])

n_clusters = 3  
for method in ['single','average','ward','complete']:
  model = AgglomerativeClustering(n_clusters=n_clusters,linkage=method)
  clusters = model.fit_predict(data)
  print(f"{method.upper()} CLUSTER")
  print("DATA PONTS | CLUSTER")
  ind=0
  for i in clusters:
    print(data[ind],"|",i)
    ind+=1
  print()
  print()


```

```python
from scipy.cluster.hierarchy import linkage,dendrogram
for method in ['single','average','centroid','complete']:
    linked = linkage(data, method=method)
    plt.figure(figsize=(10, 10))
    dendrogram(linked, orientation='top', labels=[f'Point {i+1}' for i in range(len(data))],
           distance_sort='descending', show_leaf_counts=True)
    plt.title(f'Dendrogram for Agglomerative Clustering {method.upper()}')
    plt.xlabel('Data Points')
    plt.ylabel('Euclidean Distance')
    plt.show()
```
